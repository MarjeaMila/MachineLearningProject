```python
df['gender'] = le.fit_transform(df['gender'])
df['level'] = le.fit_transform(df['level'])
df['department'] = le.fit_transform(df['department'])
df['preedusy'] = le.fit_transform(df['preedusy'])
df['infected'] = le.fit_transform(df['infected'])
df['economy_condition'] = le.fit_transform(df['economy_condition'])
df['understand'] = le.fit_transform(df['understand'])
df['communication'] = le.fit_transform(df['communication'])
df['religious'] = le.fit_transform(df['religious'])
df['obstacle'] = le.fit_transform(df['obstacle'])
df['phy_result'] = le.fit_transform(df['phy_result'])
df['on_edu_observation'] = le.fit_transform(df['on_edu_observation'])
df['psych_condition'] = le.fit_transform(df['psych_condition'])
df['phy_difficulty'] = le.fit_transform(df['phy_difficulty'])
df['on_result'] = le.fit_transform(df['on_result'])
```

```python
 y = df['psych_condition']
 x = df.drop([ 'Timestamp','psych_condition'], axis=1)
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
cm_algorithms_name =[]
true_positive = []
false_negative = []
false_positive = []
true_negative = []


algorithm_name = []
test_accuracy_data = []
cross_validation_accuracy_data = []
```

```python
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import GridSearchCV
```

## Linear Regression

```python
clf = LinearRegression()
clf.fit(x_train,y_train)
```

```python
clf.score(x_train,y_train)
```

```python
clf.score(x_test,y_test)
```

```python
cross_val_score(clf, x, y, cv=10)
```

<h1>Logistic Regression</h1>

```python
lo = LogisticRegression()
lo.fit(x_train,y_train)
```

```python
lo.score(x_train,y_train)
```

```python
lo_test_accuracy=lo.score(x_test,y_test)
lo_test_accuracy
```

```python
cross_val_score(lo, x, y, cv=10)
```

```python
lo_cross_val=cross_val_score(lo, x, y, cv=10).mean()
lo_cross_val
```

```python
y_pred = lo.predict(x_test)
y_pred
```

```python
tp, fn, fp, tn = confusion_matrix(y_test,y_pred, labels=(1,0)).ravel()
```

```python
graph = classification_report(y_test,y_pred, labels=(1,0))
print(graph)
```

```python
cm_algorithms_name.append('Logistic Regrassion')
true_positive.append(tp)
false_negative.append(fn)
false_positive.append(fp)
true_negative.append(tn)


algorithm_name.append('Logistic Regrassion')
test_accuracy_data.append(lo_test_accuracy)
cross_validation_accuracy_data.append(lo_cross_val)
```
